### Import Important Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.ar_model import AutoReg
```

### Auto Regression (AR) example

```python
# Auto Regression example - suitable for data without trend and seasonal component
# create a linear data
data = [10,12,14,16,18,20,20,22] #linear data 
```

```python
plt.figure(figsize=(10,5))
plt.plot(data,color='Red',marker='D')
plt.show()
```

```python
#Creating Auto regression (AR) model
AR_model = AutoReg(data, lags=1)
AR_model_fit = AR_model.fit() #Fitting model
```

```python
# making predictions future value between 8-20 Observation
y_pred = AR_model_fit.predict(8,12)
print(y_pred)
```

```python
pred_list = y_pred.tolist()
combinedlist = data+pred_list
```

```python
combinedlist
```

```python
# Visualization of AR Model
plt.figure(figsize=(10,5))
plt.plot(combinedlist,color='Green',Marker='*')
plt.show()
```

## Moving Average (MA) Example

```python
#suitable for data without trend and seasonality
from statsmodels.tsa.arima_model import ARMA
```

```python
data = [22,23,23,25,26.7,27,28,28,30] #linear data 
```

```python
plt.figure(figsize=(10,5))
plt.plot(data,color='Red',marker='*')
plt.show()
```

```python
# fit model
MA_model = ARMA(data, order=(0, 1)) #model with AR=0 and MA=1
MA_model_fit = MA_model.fit(disp=False)
```

```python
# make prediction
y_pred = MA_model_fit.predict(9,12)
print(y_pred)
```

```python
pred_list = y_pred.tolist()
combinedlist = data+pred_list
combinedlist
```

```python
plt.figure(figsize=(10,5))
plt.plot(combinedlist,color='Black',marker='D')
plt.show()
```
